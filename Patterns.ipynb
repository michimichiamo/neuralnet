```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.models import WheelZoomTool
import itertools

output_notebook()
```

```python
# define size of the tensors
K = 3 # number of layers
N = 10 # # number of inputs for first layer
R = 3

x_stop = [N-k*(R-1) for k in range(K)]
W_stop = [(N - (k+1)*(R-1))*R for k in range(K)]

x_size = sum(x_stop)
W_size = sum(W_stop)
y_size = N-R+1;

x_start = [0]+list(itertools.accumulate(x_stop))[:-1]
W_start = [0]+list(itertools.accumulate(W_stop))[:-1]
x_stop = list(itertools.accumulate(x_stop))

print(x_start, W_start)               
                  
# initialize addresses
x_addr = []
W_addr = []
y_addr = []

                  
# loop nest

for i in range(N-R+1):
    for j in range(R):
        W_addr.append(W_start[0] + i*R + j)
        x_addr.append(x_start[0] + i+j)
        y_addr.append(x_stop[0] + i)
```

```python
def plot_addr(size=500):
    # create three plots
    s1 = figure(plot_width=size, plot_height=size, background_fill_color="#fefefe", x_axis_label='iteration', y_axis_label='address', title='weight')
    s1.circle(range(len(W_addr)), W_addr, size=5, color="green", alpha=0.5)
    s1.toolbar.active_scroll = s1.select_one(WheelZoomTool)

    s2 = figure(plot_width=size, plot_height=size, background_fill_color="#fefefe", x_axis_label='iteration', y_axis_label='address', title='input')
    s2.triangle(range(len(x_addr)), x_addr, size=5, color="blue", alpha=0.5)
    s2.toolbar.active_scroll = s2.select_one(WheelZoomTool)

    s3 = figure(plot_width=size, plot_height=size, background_fill_color="#fefefe", x_axis_label='iteration', y_axis_label='address', title='output')
    s3.square(range(len(y_addr)), y_addr, size=5, color="red", alpha=0.5)
    s3.toolbar.active_scroll = s3.select_one(WheelZoomTool)
    
    return s1,s2,s3
```

```python
s1,s2,s3 = plot_addr()
show(row(s1, s2, s3))
```
